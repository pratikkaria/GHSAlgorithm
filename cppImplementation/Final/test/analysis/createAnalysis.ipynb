```python
import matplotlib as mpl
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt
from sklearn import preprocessing
```

```python
fp = open("logFile","r")
data = fp.read()
```

```python
data = data.split("\n\n\n\n")
```

```python
nodes = []
edges = []
messages = []
for i in data:
    temp = i.split("\n")
    numNodes = temp[0].split(":")[1]
    numEdges = temp[1].split(temp[0])[1].split(":")[1].split(" ")[1]
    numMessages = temp[3].split(":")[1].strip()
    nodes.append(int(numNodes))
    edges.append(int(numEdges))
    messages.append(int(numMessages))
```

```python
nodes = nodes / np.linalg.norm(nodes)
edges = edges/np.linalg.norm(edges)
messages = messages/np.linalg.norm(messages)
```

```python
fig = plt.figure(figsize=(8, 6))

ax = plt.axes(projection='3d')

# ax = plt.axes(projection='3d')
# ax.set_xlim3d(1200, 5000)
# ax.set_ylim3d(50,101)
# ax.set_zlim3d(3000,11000)

# Data for a three-dimensional line
zline = messages
xline = edges
yline = nodes

ax.plot3D(xline, yline, zline)

# Data for three-dimensional scattered points
zdata = messages
xdata = edges
ydata = nodes
ax.scatter3D(xdata, ydata, zdata, c=zdata, cmap='Greens');
```

```python
mpl.rcParams['legend.fontsize'] = 10

fig = plt.figure()
ax = fig.gca(projection='3d')
theta = np.linspace(-4 * np.pi, 4 * np.pi, 100)
z = messages
r = z**2 + 1
x = nodes
y = edges
ax.plot(x, y, z, label='parametric curve')
ax.legend()

plt.show()
```
